```python
import cv2
```

```python
## Method Definition
face_classifier=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
## Scale ands crop your faces
def face_extractor(img):
    ## first argument : img
    ## second argument : scalefactor(how much image size will be reduced at each image scale)
    ## third argument : how many neighors each rectangle should have
    faces = face_classifier.detectMultiScale(img, 1.3, 5)
    if faces == ():
        return None
    
    ## Crop all the faces found
    for(x,y,w,h) in faces:
        x = x-10;
        y = y-10;
        
        cropped_face = img[y:y+h+50, x:x+w+50]
    
    return cropped_face
```

```python
## Initialize webcam
cap = cv2.VideoCapture(0)
count = 0

## Collect 100 samples of my face from the webcam input
while True:
    _, frame = cap.read()
    ## Method calling
    if face_extractor(frame) is not None:
        count += 1
        face = cv2.resize(face_extractor(frame), (400, 400))
        
        ## Save file in the specified directory
        file_name_path = './Images for visualization/sudhir.' + str(count) + '.jpg'
        cv2.imwrite(file_name_path, face)
        
        
        ## put count on images and display me the live count
        cv2.putText(face, str(count), (50,50), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)
        cv2.imshow('Face Cropper', face)
    else:
        print("Face not found")
        pass
    
    if cv2.waitKey(1) == 13 or count == 100:
        break

        
cap.release()
cv2.destroyAllWindows()

print("Dataset Creation task is successfully done")
```

```python
import numpy as np
def my_label(image_name):
    name = image_name.split('.')[-3] 
    # if you have two person in your dataset
#     if name=="Ishwar":
#         return np.array([1,0])
#     elif name=="Manish":
#         return np.array([0,1])
    
    
    # if you have three person in your dataset
    if name=="sudhir":
        return np.array([1,0,0])
    elif name=="srishu":
        return np.array([0,1,0])
    elif name=="sanatan":
        return np.array([0,0,1])
```

```python
import os
from random import shuffle
from tqdm import tqdm
def my_data():
    data = []
    for img in tqdm(os.listdir("Images for visualization")):
        path=os.path.join("Images for visualization",img)
        img_data = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
        img_data = cv2.resize(img_data, (50,50))
        data.append([np.array(img_data), my_label(img)])
    shuffle(data)  
    return data
```

```python
data = my_data()
```

```python
data
```

```python
train = data[:240]  
test = data[240:]
X_train = np.array([i[0] for i in train]).reshape(-1,50,50,1)
print(X_train.shape)
y_train = [i[1] for i in train]
X_test = np.array([i[0] for i in test]).reshape(-1,50,50,1)
print(X_test.shape)
y_test = [i[1] for i in test]
```

```python
# import warnings
# warnings.filterwarnings('ignore')

import tensorflow as tf
import tflearn
from tflearn.layers.conv import conv_2d, max_pool_2d
from tflearn.layers.core import input_data, dropout, fully_connected
from tflearn.layers.estimator import regression
```

```python
tf.reset_default_graph()
convnet = input_data(shape=[50,50,1])
convnet = conv_2d(convnet, 32, 5, activation='relu')
# 32 filters and stride=5 so that the filter will move 5 pixel or unit at a time
convnet = max_pool_2d(convnet, 5)
convnet = conv_2d(convnet, 64, 5, activation='relu')
convnet = max_pool_2d(convnet, 5)
convnet = conv_2d(convnet, 128, 5, activation='relu')
convnet = max_pool_2d(convnet, 5)
convnet = conv_2d(convnet, 64, 5, activation='relu')
convnet = max_pool_2d(convnet, 5)
convnet = conv_2d(convnet, 32, 5, activation='relu')
convnet = max_pool_2d(convnet, 5)

convnet = fully_connected(convnet, 1024, activation='relu')
convnet = dropout(convnet, 0.8)
convnet = fully_connected(convnet, 3, activation='softmax')
convnet = regression(convnet, optimizer='adam', learning_rate = 0.001, loss='categorical_crossentropy')
model = tflearn.DNN(convnet, tensorboard_verbose=1)
model.fit(X_train, y_train, n_epoch=12, validation_set=(X_test, y_test), show_metric = True, run_id="FRS" )
```
